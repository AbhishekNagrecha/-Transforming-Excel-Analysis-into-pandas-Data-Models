# Using the SequenceMatcher to find Words Similarity

```python
# To compute the similarity between My name "Abhishek"  "Nagrecha"
from difflib import SequenceMatcher
m = SequenceMatcher(None, "Abhishek", "Nagrecha")
m.ratio()
```

```python
# To compute the similarity between "New Jaycob" and "New Lilianland"
from difflib import SequenceMatcher
m = SequenceMatcher(None, "New Jaycob", "New Lilianland")
m.ratio()
```

```python
# To compute the similarity between "NEW YORK METS" and "NEW YORK MEATS"
from difflib import SequenceMatcher
m = SequenceMatcher(None, "NEW YORK METS", "NEW YORK MEATS")
m.ratio()
```
